```python
rawfile=pd.read_csv("/Users/halukamaier-borst/Desktop/DW_DevelopmentalAid/Raw_Data/RAW_20yrs.csv")
```

```python
testdata=rawfile
```

```python
columns = ['A','B', 'C']
emptyframe = pd.DataFrame(columns=['Recipient','Average Sum'])
previouscell="0"
year=0
sum20=0
for index, row in testdata.iterrows():
    if row["Donor"]=="All Donors, Total":
        recipient=row["Recipient"]
        if previouscell != "0" and previouscell!=recipient:
            if year==0:
                average=0
                print(previouscell,":",sum20)
            else:
                average=sum20/year
            df2 = pd.DataFrame([[previouscell, average]], columns=["Recipient","Average Sum"])
            emptyframe=emptyframe.append(df2,ignore_index=True)
            #print(previouscell,":",sum20)
            sum20=0
            year=0
            previouscell=recipient
        else:
            sum20+=row["Value"]
            previouscell=recipient
            year+=1
        
        
```

```python
print(emptyframe.head())
```

```python
print(len(emptyframe))
emptyframedrop=emptyframe.dropna(axis=1, how='any')
print(len(emptyframedrop))
```

```python
emptyframedrop.to_csv("Raw_Data/AllDonors.csv", index=False)
```

```python
#create and empty dataframe for storing the results
emptyframe = pd.DataFrame(columns=['Recipient','Average Sum of Money'])
#set the value for the previous cell to zero as well as the year and the sum across 20 years
previouscell="0"
year=0
sum20=0
#iterate through the dataframe row by row
for index, row in testdata.iterrows():
    #if the donor is "all donors" and if the aid type is the total official
    if row["Donor"]=="All Donors, Total" and row["Aid type"]=="Total Official, Gross":
        #then set the recipient variable to recipient of the row
        recipient=row["Recipient"]
        annus=str(row["Year"])
        #this if statement is put in place to save the averages whenever the recipient changes
        #from one row to another
        #if the previouscell has not the start value of zero but is also not the current recipient
        #we have such a break
        if previouscell != "0" and previouscell!=recipient:
            #if the number of years is zero, then set the average to zero
            if year==0:
                average=0
                print(previouscell,":",sum20)
            #otherwise calculate the average by dividing the sum over the years through the year
            else:
                average=sum20/year
            #and append the name of the recipient and the average to the dataframe
            df2 = pd.DataFrame([[previouscell, average]], columns=["Recipient","Average Sum of Money"])
            emptyframe=emptyframe.append(df2,ignore_index=True)
            print(previouscell,":",sum20)
            
            try:
                #reset the sum over the years
                sum20=0
                #and add the first value to it that is weighted by population
                cashpercap=row["Value"]*1000000/(int(population.loc[recipient][annus].replace(" ", ""))*1000)
                sum20+=cashpercap
                #set the years count to zero
                year=0
                #and add one to the year count
                year+=1
                #change the previouscell to the recipient
                previouscell=recipient
            except KeyError:
                previouscell=recipient
                pass

            
        #if it is still the same recipient,...
        else:
            print(recipient, annus)
            try:
                #we add the current row value that is weighted by population
                cashpercap=row["Value"]*1000000/(int(population.loc[recipient][annus].replace(" ", ""))*1000)
                sum20+=cashpercap
                #add one year to our year count
                year+=1
                #and change the previouscell name to the current one
                previouscell=recipient
            except KeyError:
                previouscell=recipient
                pass
           
        
        
```

```python
#we first sort the dataframe by the country of birth, then the country they migrated to and then by the year
rawfilemig=rawfilemig.sort_values(by=['Country of birth/nationality', 'Country','Year'])
```

```python
#we load this data in
testdata=rawfilemig
#testdata=testdata.head(1000)
testdata
```

```python
#again we define an empty dataframe
emptyframe2 = pd.DataFrame(columns=['Country of Birth',"Destination",'Emigration'])
#set a couple of variables to default values
previousbirth="0"
previousdestination="0"
year=0
sum20=0
#and now we iterate through the dataframe
for index, row in testdata.iterrows():
    #only when the variable is about the inflow we process the data
    if row[3]=="Inflows of foreign population by nationality":
        #we set the birthcountry variable on the "Country of birth/nationality" value of the row
        birthcountry=row[1]
        annus=str(row["Year"])
        #we set the migdestination variable on the "Country" value of the row, so the country that has taken up the migrants
        migdestination=row["Country"]
        #whenever either the previousrow's country of birth or the destination is different, this if-statement is true
        if (previousbirth != "0" and previousbirth!=birthcountry) or (previousdestination != "0" and previousdestination!=migdestination) :
            #in cases we have a year value of zero, we have an average of zero
            if year==0:
                average=0
                #print(previousbirth,":",sum20)
            #otherwise the average is defined as the sum over the years divided by the number of years
            else:
                average=sum20/year
            #and then a new row is appended to the empty dataframe with the country of birth, 
            #the destination and the average number of migrants
            df2 = pd.DataFrame([[previousbirth, previousdestination,average]], columns=['Country of Birth',"Destination",'Emigration'])
            emptyframe2=emptyframe2.append(df2,ignore_index=True)
            #print(previousbirth,":",sum20)
            try:
                #the sum over the years is then resetted to one and the number of migrants from this row is added to the sum
                sum20=0
                migpercap=row["Value"]/(int(population.loc[birthcountry][annus].replace(" ", ""))*1000)

                sum20+=migpercap
                #the number of years is set to 0 and then to 1
                year=0
                year+=1
                # the previousbirth variable becomes the current birthcountry
                previousbirth=birthcountry
                # the previousdestination variable becomes the current destinationcounrty
                previousdestination=migdestination
            except KeyError:
                previousdestination=migdestination

        else:
            #if the previousbirthcountry is the same as the current one
            #as well as the previous migration destination and the current one
            #we add the number of migrants of the current row to the sum
            try:
                migpercap=row["Value"]/(int(population.loc[birthcountry][annus].replace(" ", ""))*1000)
                sum20+=migpercap
                #we add one year
                year+=1
                # the previousbirth variable becomes the current birthcountry
                previousbirth=birthcountry
                # the previousdestination variable becomes the current destinationcounrty
                previousdestination=migdestination
            except KeyError:
                previousdestination=migdestination

        
  
```
